```python
import requests
from bs4 import BeautifulSoup

url = "https://realpython.github.io/fake-jobs/"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
```

```python
jobs = soup.find_all("div", class_="column is-half")
```

```python
job_data = []

for job in jobs:
    title = job.find("h2", class_="title is-5").text.strip()
    company = job.find("h3", class_="subtitle is-6 company").text.strip()
    location = job.find("p", class_="location").text.strip()
    link = job.find("a", string="Apply")["href"]

    # # Ambil link detail job
    # link = job.find_parent("div", class_="card-footer").find(
    #     "a", string="Apply"
    # )["href"]

    job_data.append({
        "title": title,
        "company": company,
        "location": location,
        "detail_url": link
    })
```

```python
for job in job_data[:5]:
    print(job)
```

```python
import pandas as pd

df = pd.DataFrame(job_data)
df.to_csv("fake_jobs.csv", index=False)
```
